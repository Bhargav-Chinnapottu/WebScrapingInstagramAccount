```python
from bs4 import BeautifulSoup
import requests
from  fake_useragent  import UserAgent
```

```python
user_agent=UserAgent()
page=requests.get('https://www.consumerreports.org/cro/a-to-z-index/products/index.htm',headers={'user-agent':user_agent.chrome})
file_name='customer_reports.txt'
```

```python
with open('customer_reports.txt','w') as file:
    file.write(page.content.decode('utf-8')) if type(page.content)==bytes else file.write(page.content)
```

```python
def read_file():
    file=open('customer_reports.txt')
    data=file.read()
    file.close()
    return data
```

```python
soup=BeautifulSoup(read_file(),'lxml')
all_divs=soup.find_all('div',attrs={'class':'crux-body-copy'})
products=[div.a.string for div in all_divs]
for product in products:
    print(product)
```

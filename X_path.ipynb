```python
from selenium import webdriver
from time import sleep
from bs4 import BeautifulSoup
import os
import requests
import shutil
```

```python
driver=webdriver.Chrome('chromedriver')
driver.get('https://www.instagram.com')
sleep(5)

#sign_button=driver.find_element_by_xpath("//p[@class='izU2O ']/a")
#sign_button.click()

user_name_input=driver.find_element_by_xpath("//input[@aria-label='Phone number, username, or email']")
user_name_input.send_keys('__mr.cool__09')
password_input=driver.find_element_by_xpath("//input[@aria-label='Password']")
password_input.send_keys('Cbn@1437')
password_input.submit()
sleep(5)

not_now=driver.find_element_by_xpath("//button[@class='sqdOP yWX7d    y3zKF     ']")
not_now.click()
sleep(10)

turn_of=driver.find_element_by_xpath("//button[@class='aOOlW   HoLwm ']")
turn_of.click()
sleep(5)

search=driver.find_element_by_xpath("//input[@placeholder='Search']")
search.send_keys('chowdary_nandi')
target_profile='https://www.instagram.com/'+'chowdary_nandi/'
driver.get(target_profile)

no_of_posts=driver.find_element_by_xpath("//span[@class='g47SY ']")
no_of_posts=str(no_of_posts.text).replace(',','')
no_of_posts=int(no_of_posts)
if no_of_posts>12:
    no_of_scrolls=int(no_of_posts/12)+3
    
    for value in range(no_of_scrolls):
        driver.execute_script('window.scrollTo(0, document.body.scrollHeight);')
        sleep(1)
        
#login_button=driver.find_element_by_xpath('//button[@class="sqdOP  L3NKy   y3zKF     "]')
#login_button.click()

path='/Users/USER/Desktop/WebScraping/instaphotos'
if not os.path.exists(path):
    os.mkdir(path)
    
soup=BeautifulSoup(driver.page_source,'lxml')
all_images=soup.find_all('img',class_="FFVAD")
print(len(all_images))
for index,image in enumerate(all_images):
    filename='image_'+str(index)+'.jpg'
    image_path=os.path.join(path,filename)
    link=image['src']
    response=requests.get(link,stream=True)
    with open(image_path,'wb') as file:
        shutil.copyfileobj(response.raw,file)

captions_folder_path=os.path.join(path,'captions')
if not os.path.exists(captions_folder_path):
    os.mkdir(captions_folder_path)
for index,image in enumerate(all_images):
    try:
        caption=image['alt']
    except KeyError:
        caption='No caption exixts for this image'
    filename='caption_'+str(index)+'.txt'
    filepath=os.path.join(captions_folder_path,filename)
    link=image['src']
    with open(filepath,'wb') as file:
        file.write(str('link:'+str(link)+'\n'+'caption:'+caption).encode())
        

sleep(15)
driver.close()
```
